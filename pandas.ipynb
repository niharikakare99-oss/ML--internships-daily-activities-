### Concate Data Frames

```python
india_weather=pd.DataFrame({
    "city":["kakinada","hyderabad","vizag"],
    "temparature":[32,45,35]
})
india_weather
```

```python
US_weather=pd.DataFrame({
    "city":["newyork","Chicago","orland"],
    "temparature":[45,32,33]
})
US_weather
```

```python
df=pd.concat([india_weather,US_weather])
df
```

```python
a=pd.concat([india_weather,US_weather],ignore_index=True)##ignore index=give the index line by line 012345
a
```

```python
df=pd.concat([india_weather,US_weather],axis=1)##axis=1> it gives the index side by side
df
```

### merge data frames

```python
temp_df=pd.DataFrame({
    "city":["hyd","mumbai","kkd","vizag"],
    "temparature":[32,43,29,30]
})
temp_df
```

```python
humidity_df=pd.DataFrame({
    "city":["hyd","kirlampudi","prathipadu","vizag"],
    "humidity":[32,43,29,30]
})
humidity_df
```

```python
df=pd.merge(temp_df,humidity_df,on="city")
df
```

```python
df=pd.merge(temp_df,humidity_df,how="outer")
df
```

```python
df.isnull()## to find the missing values or null values
```

```python
df.isnull().sum()
```

```python
df['temparature']=df['temparature'].fillna(df.temparature.min())
```
