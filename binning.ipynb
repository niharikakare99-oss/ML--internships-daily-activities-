```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.DataFrame({
    "age":[18,23,41,50,34,55,21,19]
})
data
```

### Equal Width

```python
data["age_bin"]=pd.cut(data["age"],bins=4)
data
```

```python
plt.hist(data["age"],bins=4)
plt.xlabel("age")
plt.ylabel("frequency")
plt.title("equal_width binning using pd.cut()")
plt.show()
```

### Equal Frequency

```python
data=pd.DataFrame({
     "age":[18,23,41,50,34,55,21,19]
})
data
```

```python
data["age_bin"]=pd.qcut(data["age"],q=2)
data
```

```python
plt.hist(data["age"],bins=2)
plt.xlabel("age")
plt.ylabel("frequency")
plt.title("equal_width binning using pd.qcut()")
plt.show()
```
