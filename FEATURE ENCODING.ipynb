##### FEATURE ENCODING

##### feature encoding is the process of converting a non numerical data into a numerical format that machine learning algorithms can understand and process.
###### we choose this because...... the type of categorical data, the number of unique values in a categorical column,and the mode we plan to train types of categorical data.... nominal(no order), ordinal order(order)
###### NOMINAL DATA: no order existence ex:red, ordinal data order exists ex: size(small<medium<large)
                     rating(low<medium<high) importance: text data can't be use directly in model training. proper encoding: 
                     improves model accuracy prevents wrong assumptions make training stable
### common feature encoding technique(7)
     Label encoding, one hot encoding, target encoding, ordinal encoding, frequency encoding, binary encoding, feature hasmap
#### LABEL ENCODING:
     each category is asssigned a unique integer (eg: "red":0,"green":1).
     use for ordinal data (inherent order) or with tree based models that are nots sensitive to that numerical order. 


```python
import pandas as pd
df=pd.DataFrame({
    "color":["red","green","blue","green"],
    "size":["small","medium","large","medium"],
    "city":["hyd","delhi","mumbai","delhi"],
    "target":[10,20,15,25]
})
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df["color_label"]= LabelEncoder().fit_transform(df["color"])
df[["color","color_label"]]
```

#### ONE-HOT CODING:
      1-present,0-absent
      create a new binary column for each category, with a 1 indicating the presence of that category and 0 otherwise.
      ideal for nominal data with low to medium cardinalitym and for linear models and nueral networks.

```python
one_hot=pd.get_dummies(df["color"],prefix="color") # if the value is there then it will be
one_hot
```

```python
oh=pd.get_dummies(df,columns=["color"])
oh
```

### ordinal encoding:
     Similar to label encoding, but the integer assignment is based on the specific, predefined order or rank of the categories(eg: "low":0,"medium":1,"high":2)
     use specifically when the categorical variable ahs the natural meaning ful order.

```python
from sklearn.preprocessing import OrdinalEncoder # takes all the values based on order
```

```python
order=[["small","medium","large"]]
df[["size_ordinal"]]= OrdinalEncoder(categories=order).fit_transform(df[["size"]])
df[["size","size_ordinal"]]
```

### Target Encoding 
     Each category is replaced with the mean of the target variable for that category.
     Effective for high-cardinality features and when there's a strong relationship bettween the category and the target       variable, but requires careful validation to valid target values

```python
te=df.groupby("city")["target"].mean()
df["city_target_Enc"]=df["city"].map(te)
df[["city", "city_target_Enc"]]
```
